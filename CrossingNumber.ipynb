```python
import random

def are_more_than_three_collinear(points):
    n = len(points)
    if n < 3:
        return False  # Fewer than 3 points can't be collinear

    for i in range(n):
        slopes = {}  # Dictionary to store slopes
        for j in range(n):
            if i != j:
                dx = points[j][0] - points[i][0]
                dy = points[j][1] - points[i][1]
                if dx == 0:
                    s = float('inf')  # Represent vertical slope
                else:
                    s = dy / dx
                slopes[s] = slopes.get(s, 0) + 1

        # Check if any slope has more than 2 occurrences
        if any(count >= 2 for count in slopes.values()):
            return True
    return False


def generate_non_collinear_points(n, min_val=-100, max_val=100):
    points = []
    
    while len(points) < n:
        x, y = random.randint(min_val, max_val), random.randint(min_val, max_val)
        # Temporarily add the new point
        temp_points = points + [(x, y)]
        # Check if this point makes any three points collinear
        if not are_more_than_three_collinear(temp_points):  
            points.append((x, y))
    
    return points

import matplotlib.pyplot as plt
# Define the function to plot points
def plot_points(points):
    x_coords = [p[0] for p in points]
    y_coords = [p[1] for p in points]
    
    plt.figure(figsize=(8, 8))
    plt.scatter(x_coords, y_coords, color='blue', label='Points')
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    plt.grid(color='gray', linestyle=':', linewidth=0.5)
    plt.title('Generated Non-Collinear Points')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.show()

# Generate points
points = generate_non_collinear_points(50)
# Plot the points
plot_points(points)



```
